```python
movie_names = ["SIMPS_9a_c"]#["MARCH_6b_c", "SIMPS_3a_c", "MISSP_9a_c"]# ["DEEPB_11a_c"]#["DEEPB_5b_c"]#["Cloud_13a_c", "ADVEN_12a_c", "DEEPB_9a_c"]
```

```python
for movie_name in tqdm(movie_names):
    # get movie data path
    print(movie_name)

    # get movie data
    cap = cv2.VideoCapture(f"../Clips_small_size/{movie_name}.mov")
    
    # make save directory
    save_dir = f"../preprocessed_data/clips_png_original/{movie_name}"
    os.makedirs(save_dir, exist_ok=True)

    # save frames
    count = 0
    while True:
        ret, frame = cap.read()
        if ret == True:
            count += 1
            cv2.imwrite(save_dir + '/{:04d}.png'.format(count), frame)
            #cv2.imwrite(save_dir + '/{:04d}.png'.format(count), frame)
        else:
            print(count)
            break
```
